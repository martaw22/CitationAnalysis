```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
firstyear = pd.read_csv('author_avgcitations_groupbyyearoffirstpublication_method2.csv')
byfield = pd.read_csv('citations_byfield_method2.csv')
perjournal = pd.read_csv('pub_aveperjournal_method2.csv')
byyear = pd.read_csv('topcited_citations_byyear.csv')
delay = pd.read_csv('delay_recognition.csv')
perjournal_topfifty = pd.read_csv('perjournal_top50.csv')
```

```python
#perjournal_topfifty = perjournal_topfifty.drop([0])
perjournal_topfifty = perjournal_topfifty[:25]
```

```python
perjournal_topfifty = perjournal_topfifty.replace('Proceedings of the National Academy of Sciences of the United States of America', 'PNAS')
perjournal_topfifty = perjournal_topfifty.replace('The Journal of neuroscience : the official journal of the Society for Neuroscience', 'Journal of Neuroscience')
```

```python
index = list(range(0,25))
```

```python
perjournal_topfifty['index'] = index
```

```python

ax = perjournal_topfifty.plot.bar(x='index', y='AVG(RATIO)', title='Average Self-Citation Rate' + '\n' 'per Publication per Journal', fontsize = 11)
plt.tight_layout()
plt.ylabel('Self-Citation Rate')
plt.xlabel('Journal')
ax.get_legend().remove()
plt.tight_layout()
plt.rcParams.update({'font.size': 14})
plt.savefig('perjournal_output.pdf', bbox_inches = 'tight')
```

```python
#hist = delay.hist(column='DELAY', title = 'Number of Years until an Author has Maximum Citations in one Year')
plt.hist(delay['DELAY'])
plt.title('Number of Years until a Paper has Maximum Citations in one Year')
plt.ylabel('Count of Papers')
plt.xlabel('Number of Years')
plt.rcParams.update({'font.size': 16})
plt.savefig('numyears_untilmax_output.pdf')
plt.show()
```

```python
plt.plot(delay['DELAY'], delay['TOTAL_CITATIONS'], 'o')
plt.ylabel('Total Number of Citations')
plt.xlabel('Number of Years until Max Citations')
plt.title('Delayed Recognition - Number of Years until a Paper has the Maximum Number of Citations in a Year')
plt.savefig('delayedrec_numyears_output.pdf', bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(delay['FIRSTYEAR'], delay['DELAY'], 'o')
plt.ylabel('Number of Years until Max Citations')
plt.xlabel('Year of Publication')
plt.title('Number of Years until a Paper has the Maximum Number of Citations in a Year')
plt.savefig('delayedrec_output.pdf')
plt.show()
```

```python
byfield = byfield.sort_values(by=['AVG(RATIO)'])
ax = byfield.plot.bar(x='FIELDOFSTUDY', y='AVG(RATIO)', title='Average Self-Citation Rate' + '\n' + 'per Author per Field', fontsize = 11)
plt.ylabel('Self-Citation Rate')
plt.xlabel('Field of Study')
plt.rcParams.update({'font.size': 14})
ax.get_legend().remove()
plt.tight_layout()
plt.savefig('perfield_output.pdf')
```

```python
plt.scatter(byfield['AVG(ALL_CITE)'], byfield['AVG(SELF_CITE)'], s=byfield['AVG(RATIO)'], data=byfield)
plt.ylabel('Average Number of Self-Citations per Author in a Given Field')
plt.xlabel('Average Number of Citations per Author in a Given Field')
plt.title('Relationship between Citation and Self-Citation by Field of Study')
plt.show()
```

```python
plt.plot(firstyear['MINYEAR'], firstyear['AVG(RATIO)'], 'go')
plt.ylabel('Self-Citation Rate')
plt.xlabel('Year of First Publication')
plt.title('Average Self-Citation Rate for Authors who' + '\n' + 'Published Their First Paper in a Given Year')
plt.tight_layout()
plt.rcParams.update({'font.size': 16})
plt.savefig('selfcitationsperyear_output.pdf', bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(firstyear['MINYEAR'], firstyear['AVG(ALL_CITE)'], 'o', label='Total Citations')
plt.plot(firstyear['MINYEAR'], firstyear['AVG(SELF_CITE)'], 'mo', label='Self-Citations')
plt.ylabel('Average Citation Counts')
plt.xlabel('Year of First Publication')
plt.title('Citation Counts for Authors who Published' + '\n' + 'Their First Paper in a Given Year')
plt.legend(prop={'size': 10})
plt.tight_layout()
plt.rcParams.update({'font.size': 16})
plt.savefig('citationsperyear_output.pdf', bbox_inches = 'tight')
plt.show()
```
